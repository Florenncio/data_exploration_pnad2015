```python
import pandas as pd
import seaborn as sns
```

```python
dados = pd.read_csv("dados/dados.csv")
```

```python
ax = sns.boxplot(data=dados.query('(UF==35 or UF==29) and (Renda < 10000)'), 
                 x='Renda', y='UF', orient='h')
ax.figure.set_size_inches(15,8)
ax.set_title('Renda(R$) | Bahia x São Paulo', fontsize=14)
ax.set_xlabel('R$', fontsize=14)
ax.set_ylabel('Unidade Federativa \n(35 = São Paulo | 29 = Bahia)', fontsize=14)
ax
```
