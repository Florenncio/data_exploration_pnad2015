```python
import pandas as pd
import seaborn as sns
```

```python
dados = pd.read_csv('dados/dados.csv')
```

```python
dados.head()
```

```python
df = dados.query('UF >= 21 and  UF <= 29 and Renda < 10000')
```

```python
df.UF = df.UF.map({
    21:'Maranhão',
    22:'Piauí',
    23:'Ceará',
    24:'Rio Grande do Norte',
    25:'Paraíba',
    26:'Pernambuco',
    27:'Alagoas',
    28:'Sergipe',
    29:'Bahia'
})
```

```python
ax = sns.boxplot(data=df, x='Renda', y='UF')
ax.figure.set_size_inches(15,8)
ax.set_title('Rendimento Mensal do Chefe de Familia na Região Nordeste (Menores que R$ 10.000,00)', fontsize=14)
ax.set_xlabel('R$', fontsize=14)
ax.set_ylabel('Estado', fontsize=14)
ax
```

```python
ax = sns.boxplot(data=dados.query('(UF==25) and (Renda < 10000)'), 
                 x='Renda', y='Sexo', orient='h')
ax.figure.set_size_inches(15,8)
ax.set_title('Rendimento Mensal do Chefe de Familia no Estado da Paraíba (Menores que R$ 10.000,00) ', fontsize=14)
ax.set_xlabel('R$', fontsize=14)
ax.set_ylabel('Sexo \n(0 = Masculino | 1 = Feminino))', fontsize=14)
ax
```
